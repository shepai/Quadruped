```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
motors=np.load("/its/home/drs25/Documents/GitHub/Quadruped/Code/GAs/motors_7.npy")
print(motors.shape)
#motors=np.radians(motors)
```

```python
plt.plot(motors[:,0],motors[:,1],label="leg1")
plt.plot(motors[:,3],motors[:,4],label="leg2")
plt.plot(motors[:,6],motors[:,7],label="leg3")
plt.plot(motors[:,8],motors[:,9],label="leg4")
plt.xlabel("Ankle degrees")
plt.ylabel("Knee degrees")
plt.title("Leg positions")
plt.legend(loc="lower right")
plt.show()

oscillations = np.diff(motors[0:,3])
zero_crossings = np.where(np.diff(np.sign(oscillations)) != 0)[0] + 1  # +1 to correct index shift

# Compute phase lengths (time steps between zero-crossings)
phase_lengths = np.diff(zero_crossings)

print("Zero-crossing indices:", zero_crossings)
print("Phase lengths:", phase_lengths)
```
